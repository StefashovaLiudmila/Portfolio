- **–§–∏–ª—å—Ç—Ä–∞—Ü–∏—è**:  
  - –¢–æ–ª—å–∫–æ **–≥–æ—Ä–æ–¥—Å–∫–∏–µ –æ–±—ä—è–≤–ª–µ–Ω–∏—è** (`TYPE = '–≥–æ—Ä–æ–¥'`),  
  - –£–¥–∞–ª–µ–Ω—ã **–≤—ã–±—Ä–æ—Å—ã** –ø–æ –ø–ª–æ—â–∞–¥–∏, –∫–æ–º–Ω–∞—Ç–∞–º, –±–∞–ª–∫–æ–Ω–∞–º, –≤—ã—Å–æ—Ç–µ –ø–æ—Ç–æ–ª–∫–æ–≤ (99-–π –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å),  
  - –û–±—ä—è–≤–ª–µ–Ω–∏—è —Å `days_exposition IS NULL` –æ—Ç–Ω–µ—Å–µ–Ω—ã –∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ **¬´–¥—Ä—É–≥–∏–µ¬ª** ‚Äî –Ω–µ –∏—Å–∫–ª—é—á–µ–Ω—ã –∏–∑ –∞–Ω–∞–ª–∏–∑–∞.  
- **–ú–µ—Ç—Ä–∏–∫–∏**:  
  - –°—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –∑–∞ –º¬≤,  
  - –ú–µ–¥–∏–∞–Ω–∞ –ø–æ –∫–æ–º–Ω–∞—Ç–∞–º/–±–∞–ª–∫–æ–Ω–∞–º/—ç—Ç–∞–∂–∞–º,  
  - –î–æ–ª—è –ø—Ä–æ–¥–∞–∂, —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—å, —Å–∫–æ—Ä–æ—Å—Ç—å —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏.

```python
# –ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
import pandas as pd
```

## –ó–∞–¥–∞—á–∞ 1. –í—Ä–µ–º—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏–π

–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞–∏–±–æ–ª–µ–µ –∏ –Ω–∞–∏–º–µ–Ω–µ–µ –ª–∏–∫–≤–∏–¥–Ω—ã—Ö —Å–µ–≥–º–µ–Ω—Ç–æ–≤ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏ –≤ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ –∏ –õ–µ–Ω–∏–Ω–≥—Ä–∞–¥—Å–∫–æ–π –æ–±–ª–∞—Å—Ç–∏ –ø–æ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏–π.

```python
# SQL-–∑–∞–ø—Ä–æ—Å, –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω—ã–π –≤ DBeaver
sql_task1 = """
-- –û–ø—Ä–µ–¥–µ–ª–∏–º –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (–≤—ã–±—Ä–æ—Å—ã) –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–µ–π:
WITH limits AS (
    SELECT  
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY total_area) AS total_area_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY rooms) AS rooms_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY balcony) AS balcony_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY ceiling_height) AS ceiling_height_limit_h,
        PERCENTILE_DISC(0.01) WITHIN GROUP (ORDER BY ceiling_height) AS ceiling_height_limit_l
    FROM real_estate.flats     
),
-- –ù–∞–π–¥—ë–º id –æ–±—ä—è–≤–ª–µ–Ω–∏–π, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤—ã–±—Ä–æ—Å—ã:
filtered_id AS(
    SELECT id
    FROM real_estate.flats  
    WHERE 
        total_area < (SELECT total_area_limit FROM limits)
        AND (rooms < (SELECT rooms_limit FROM limits) OR rooms IS NULL)
        AND (balcony < (SELECT balcony_limit FROM limits) OR balcony IS NULL)
        AND ((ceiling_height < (SELECT ceiling_height_limit_h FROM limits)
            AND ceiling_height > (SELECT ceiling_height_limit_l FROM limits)) OR ceiling_height IS NULL)
    ), 
-- –ì–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è
categories AS (
	SELECT 
		fi.id AS id,
		days_exposition,
		last_price,
		last_price/total_area AS avg_pr_sm,
		total_area,
		rooms,
		balcony,
		floor,
		kitchen_area,
		ceiling_height,
		is_apartment,
		parks_around3000,
		ponds_around3000,
		open_plan,
		airports_nearest,
		CASE 
			WHEN city = '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥'
			THEN '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥'
			ELSE '–õ–µ–Ω–û–±–ª'
			END AS region,
		CASE
			WHEN days_exposition IS NULL 
			THEN '–¥—Ä—É–≥–∏–µ'
			WHEN days_exposition <= 30
			THEN '  –¥–æ –º–µ—Å—è—Ü–∞'
			WHEN days_exposition <= 90
			THEN '  –¥–æ —Ç—Ä–µ—Ö –º–µ—Å—è—Ü–µ–≤'
			WHEN days_exposition <= 180
			THEN ' –¥–æ –ø–æ–ª—É–≥–æ–¥–∞'
			ELSE '–±–æ–ª–µ–µ –ø–æ–ª—É–≥–æ–¥–∞'
			END AS activity_segment
	FROM filtered_id AS fi
	INNER JOIN real_estate.flats AS f ON fi.id = f.id
	INNER JOIN real_estate.advertisement AS a ON a.id = f.id
	INNER JOIN real_estate.city AS c USING (city_id)
	INNER JOIN real_estate.TYPE AS t ON f.type_id = t.type_id
	WHERE TYPE = '–≥–æ—Ä–æ–¥' AND first_day_exposition BETWEEN '2015-01-01' and '2018-12-31' 
	)
SELECT 
	region,
	activity_segment,
	count(id) AS ad_count, -- –∫–æ–ª–∏—á–µ—Ç—Å–≤–æ –æ–±—ä—è–≤–ª–µ–Ω–∏–π
	(count(id)::float*100/sum(count(id)) OVER (PARTITION BY region))::numeric(10,1) AS share, -- –¥–æ–ª—è –æ–±—ä—è–≤–ª–µ–Ω–∏–π –≤ —Ä–∞–∑—Ä–µ–∑–µ —Ä–µ–≥–∏–æ–Ω–∞ –≤ –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö
	ROUND(AVG(avg_pr_sm::NUMERIC),0) AS avg_pr_sm, -- —Å—Ä–µ–¥–Ω—è—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞
	ROUND(avg(total_area)::numeric,1) AS avg_area, -- —Å—Ä–µ–¥–Ω—è—è –ø–ª–æ—â–∞—Ç—å –∫–≤–∞—Ä—Ç–∏—Ä—ã
	ROUND(AVG(kitchen_area)::NUMERIC,2) AS avg_kitchen_area, -- —Å—Ä–µ–¥–Ω—è—è –ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏
	ROUND(AVG(ceiling_height)::NUMERIC,2) AS avg_ceil_height, -- —Å—Ä–µ–¥–Ω—è—è –≤—ã—Å–æ—Ç–∞ –ø–æ—Ç–æ–ª–∫–æ–≤
	PERCENTILE_DISC(0.5) WITHIN GROUP (ORDER BY rooms) AS med_room, -- –º–µ–¥–∏–∞–Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∫–æ–º–Ω–∞—Ç
	PERCENTILE_DISC(0.5) WITHIN GROUP (ORDER BY balcony) AS med_balcony, -- –º–µ–¥–∏–∞–Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –±–∞–ª–∫–æ–Ω–æ–≤
	PERCENTILE_DISC(0.5) WITHIN GROUP (ORDER BY floor) AS med_floor, -- –º–µ–¥–∏–∞–Ω–∞ —ç—Ç–∞–∂–∞ –∫–≤–∞—Ä—Ç–∏—Ä—ã
	PERCENTILE_DISC(0.5) WITHIN GROUP (ORDER BY parks_around3000) AS med_parks, -- –º–µ–¥–∏–∞–Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –ø–∞—Ä–∫–æ–≤ –≤ —Ä–∞–¥–∏—É—Å–µ —Ç—Ä–µ—Ö –∫–∏–ª–æ–º–µ—Ç—Ä–æ–≤
	PERCENTILE_DISC(0.5) WITHIN GROUP (ORDER BY ponds_around3000) AS med_ponds, -- –º–µ–¥–∏–∞–Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –≤–æ–¥–æ–µ–º–æ–≤ –≤ —Ä–∞–¥–∏—É—Å–µ —Ç—Ä–µ—Ö –∫–∏–ª–æ–º–µ—Ç—Ä–æ–≤
	round(sum(is_apartment)::numeric/count(is_apartment)::NUMERIC*100,2) AS per_apartment,  -- –ø—Ä–æ—Ü–µ–Ω—Ç –∫–≤–∞—Ä—Ç–∏—Ä, —è–≤–ª—è—é—â–∏—Ö—Å—è –∞–ø–∞—Ä—Ç–∞–º–µ–Ω—Ç–∞–º–∏
	round(sum(open_plan)::numeric/count(open_plan)::NUMERIC*100,2) AS per_open_pl,  -- –ø—Ä–æ—Ü–µ–Ω—Ç –∫–≤–∞—Ä—Ç–∏—Ä —Å –æ—Ç–∫—Ä—ã—Ç–æ–π –ø–ª–∞–Ω–∏—Ä–æ–≤–∫–æ–π
	ROUND(AVG(airports_nearest)::NUMERIC,0) AS avg_airport_dist -- —Å—Ä–µ–¥–Ω–µ–µ —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ –¥–æ –±–ª–∏–∂–∞–π—à–µ–≥–æ –∞—ç—Ä–æ–ø–æ—Ä—Ç–∞
FROM categories
GROUP BY 
	region,
	activity_segment
ORDER BY 
	region DESC,
	activity_segment;
"""

pd.read_csv("task1_result.csv", encoding="utf-8")
```

- **–õ–û**: –≤ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç–µ **–ø—Ä–µ–º–∏—É–º-—Å–µ–≥–º–µ–Ω—Ç** (–Ω–µ—Å–º–æ—Ç—Ä—è –Ω–∞ –º–µ–Ω—å—à–∏–π –æ–±—ä—ë–º), –æ—Å–æ–±–µ–Ω–Ω–æ –≤ –ú—É—Ä–∏–Ω–æ, –ö—É–¥—Ä–æ–≤–æ, –®—É—à–∞—Ä–∞—Ö.

## –ó–∞–¥–∞—á–∞ 2. –°–µ–∑–æ–Ω–Ω–æ—Å—Ç—å –æ–±—ä—è–≤–ª–µ–Ω–∏–π

–ê–Ω–∞–ª–∏–∑ —Å–µ–∑–æ–Ω–Ω—ã—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏: –∫–æ–≥–¥–∞ —á–∞—â–µ –≤—Å–µ–≥–æ –ø—É–±–ª–∏–∫—É—é—Ç –∏ —Å–Ω–∏–º–∞—é—Ç (–ø—Ä–æ–¥–∞—é—Ç) –æ–±—ä—è–≤–ª–µ–Ω–∏—è.

```python
# SQL-–∑–∞–ø—Ä–æ—Å, –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω—ã–π –≤ DBeaver
sql_task2 = """
-- –û–ø—Ä–µ–¥–µ–ª–∏–º –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (–≤—ã–±—Ä–æ—Å—ã) –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–µ–π:
WITH limits AS (
    SELECT  
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY total_area) AS total_area_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY rooms) AS rooms_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY balcony) AS balcony_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY ceiling_height) AS ceiling_height_limit_h,
        PERCENTILE_DISC(0.01) WITHIN GROUP (ORDER BY ceiling_height) AS ceiling_height_limit_l
    FROM real_estate.flats     
),
-- –ù–∞–π–¥—ë–º id –æ–±—ä—è–≤–ª–µ–Ω–∏–π, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤—ã–±—Ä–æ—Å—ã:
filtered_id AS(
    SELECT id
    FROM real_estate.flats  
    WHERE 
        total_area < (SELECT total_area_limit FROM limits)
        AND (rooms < (SELECT rooms_limit FROM limits) OR rooms IS NULL)
        AND (balcony < (SELECT balcony_limit FROM limits) OR balcony IS NULL)
        AND ((ceiling_height < (SELECT ceiling_height_limit_h FROM limits)
            AND ceiling_height > (SELECT ceiling_height_limit_l FROM limits)) OR ceiling_height IS NULL)
),
-- –í—ã–¥–µ–ª—è–µ–º –º–µ—Å—è—Ü –∏–∑ –¥–∞—Ç—ã –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –∏ –¥–∞—Ç—ã —Å–Ω—è—Ç–∏—è
seasonality AS (
	SELECT 
		EXTRACT('month' FROM first_day_exposition) AS month_first,
		EXTRACT('month' FROM (first_day_exposition + INTERVAL '1 day' * days_exposition::integer)) AS month_end,
		first_day_exposition,
		days_exposition,
		last_price,
		last_price/total_area AS avg_pr_sm,
		total_area
	FROM filtered_id AS fi
	INNER JOIN real_estate.flats AS f ON fi.id = f.id
	INNER JOIN real_estate.advertisement AS a ON a.id = f.id
	INNER JOIN real_estate.city AS c USING (city_id)
	INNER JOIN real_estate.TYPE AS t ON f.type_id = t.type_id
	WHERE TYPE = '–≥–æ—Ä–æ–¥' AND first_day_exposition BETWEEN '2015-01-01' and '2018-12-31'
),
-- —Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø–æ –¥–∞—Ç–µ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏–π
publication_activity AS (
    SELECT 
    	month_first,
        COUNT(*) AS count_pub,
        sum(count(*)) OVER () AS total_count,
        RANK() OVER (ORDER BY COUNT(*) DESC) AS pub_rank,
        ROUND(AVG(avg_pr_sm::NUMERIC),0) AS avg_pr_sm_pub,
        ROUND(AVG(total_area)::NUMERIC,2) AS avg_area_pub
    FROM seasonality
    GROUP BY month_first
),
-- —Å—á–∏—Ç–∞–µ–º –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –ø–æ –¥–∞—Ç–µ —Å–Ω—è—Ç–∏—è –æ–±—ä—è–≤–ª–µ–Ω–∏–π
removal_activity AS (
    SELECT 
        month_end,
        COUNT(*) AS count_rem,
        sum(count(*)) OVER () AS total_count_rem,
        DENSE_RANK() OVER (ORDER BY COUNT(*) DESC) AS rem_rank,
        ROUND(AVG(avg_pr_sm::NUMERIC),0) AS avg_pr_sm_rem,
        ROUND(AVG(total_area)::NUMERIC,2) AS avg_area_rem
    FROM seasonality
    WHERE days_exposition IS NOT NULL 
    GROUP BY month_end
)
-- —Ñ–æ—Ä–º–∏—Ä—É–µ–º –∏—Ç–æ–≥–æ–≤—É—é —Ç–∞–±–ª–∏—Ü—É
SELECT 
    CASE 
    	WHEN pa.month_first = 1 THEN '—è–Ω–≤'
    	WHEN pa.month_first = 2 THEN '—Ñ–µ–≤'
    	WHEN pa.month_first = 3 THEN '–º–∞—Ä'
    	WHEN pa.month_first = 4 THEN '–∞–ø—Ä'
    	WHEN pa.month_first = 5 THEN '–º–∞–π'
    	WHEN pa.month_first = 6 THEN '–∏—é–Ω'
    	WHEN pa.month_first = 7 THEN '–∏—é–ª'
    	WHEN pa.month_first = 8 THEN '–∞–≤–≥'
    	WHEN pa.month_first = 9 THEN '—Å–µ–Ω'
    	WHEN pa.month_first = 10 THEN '–æ–∫—Ç'
    	WHEN pa.month_first = 11 THEN '–Ω–æ—è'
    	WHEN pa.month_first = 12 THEN '–¥–µ–∫'
    END AS month_pupl, -- –º–µ—Å—è—Ü –ø—É–±–ª–∏–∫–∞—Ü–∏–∏
    pub_rank, -- —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏–µ –º–µ—Å—è—Ü—É –ø—É–±–ª–∏–∫–∞—Ü–∏–∏
    count_pub, -- –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø—É–±–ª–∏–∫–æ–≤–∞–Ω–Ω—ã—Ö –æ–±—ä—è–≤–ª–µ–Ω–∏–π
    ROUND(100*count_pub/total_count::numeric,2) AS avg_per_pub,
    avg_pr_sm_pub, -- —Å—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –∑–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –º–µ—Ç—Ä
    avg_area_pub, -- —Å—Ä–µ–¥–Ω—è—è –ø–ª–æ—â–∞–¥—å –∫–≤–∞—Ä—Ç–∏—Ä
    rem_rank, -- —Ä–∞–Ω–∂–∏—Ä–æ–≤–∞–Ω–∏–µ –ø–æ –º–µ—Å—è—Ü—É –ø—Ä–æ–¥–∞–∂–∏
    count_rem, -- –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–≤–∞—Ä—Ç–∏—Ä
    ROUND(100*count_rem/total_count_rem::numeric,2) AS avg_per_rem,
    avg_pr_sm_rem, -- —Å—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –∑–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –º–µ—Ç—Ä –ø—Ä–∏ –ø—Ä–æ–¥–∞–∂–µ
    avg_area_rem -- —Å—Ä–µ–¥—Ä—è—è –ø–ª–æ—â–∞—Ç—å –ø—Ä–æ–¥–∞–Ω–Ω—ã—Ö –∫–≤–∞—Ä—Ç–∏—Ä
FROM publication_activity AS pa
LEFT JOIN removal_activity AS ra ON pa.month_first = ra.month_end
ORDER BY pa.month_first;
"""

pd.read_csv("task2_result.csv", encoding="utf-8")
```

## –í—ã–≤–æ–¥—ã –ø–æ –ó–∞–¥–∞—á–µ 2. –°–µ–∑–æ–Ω–Ω–æ—Å—Ç—å –æ–±—ä—è–≤–ª–µ–Ω–∏–π

### üîπ –ü–∏–∫ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏
- **–ü—É–±–ª–∏–∫–∞—Ü–∏—è**: –Ω–æ—è–±—Ä—å (1 569), –æ–∫—Ç—è–±—Ä—å (1 437), —Ñ–µ–≤—Ä–∞–ª—å (1 369)  
- **–ü—Ä–æ–¥–∞–∂–∏ (—Å–Ω—è—Ç–∏–µ)**: –æ–∫—Ç—è–±—Ä—å (1 360), –Ω–æ—è–±—Ä—å (1 301), —Å–µ–Ω—Ç—è–±—Ä—å (1 238)  
‚Üí **–û—Å–µ–Ω—å (—Å–µ–Ω—Ç—è–±—Ä—å‚Äì–Ω–æ—è–±—Ä—å)** ‚Äî –ø–∏–∫ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è –∏ —Å–ø—Ä–æ—Å–∞.

### üîπ –ù–∏–∑–∫–∏–π —Å–µ–∑–æ–Ω
- **–ú–∞–π** ‚Äî –º–∏–Ω–∏–º—É–º –∏ –ø–æ –ø—É–±–ª–∏–∫–∞—Ü–∏—è–º (891), –∏ –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º (729).  
- **–í–µ—Å–Ω–∞** ‚Äî –Ω–∞–∏–º–µ–Ω–µ–µ –∞–∫—Ç–∏–≤–Ω—ã–π –ø–µ—Ä–∏–æ–¥.

### üîπ –¶–µ–Ω—ã –∏ –ø–ª–æ—â–∞–¥—å
- **–¶–µ–Ω—ã —Å—Ç–∞–±–∏–ª—å–Ω—ã**: 102‚Äì105 —Ç—ã—Å. ‚ÇΩ/–º¬≤, –Ω–µ–±–æ–ª—å—à–æ–π —Ä–æ—Å—Ç –≤ —è–Ω–≤–∞—Ä–µ –∏ –∞–≤–≥—É—Å—Ç–µ.
- **–ü–ª–æ—â–∞–¥—å –ª–µ—Ç–æ–º –º–µ–Ω—å—à–µ**: –∞–≤–≥—É—Å—Ç ‚Äî 56.8 –º¬≤, —Å–µ–Ω—Ç—è–±—Ä—å ‚Äî 61.0 –º¬≤.

### üîπ –†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏
- **–ú–∞—Ä–∫–µ—Ç–∏–Ω–≥ –∏ –∑–∞–ø—É—Å–∫ –æ–±—ä–µ–∫—Ç–æ–≤** ‚Äî **–∞–≤–≥—É—Å—Ç‚Äì—Å–µ–Ω—Ç—è–±—Ä—å**.
- **–ò–∑–±–µ–≥–∞—Ç—å –∫—Ä—É–ø–Ω—ã—Ö –∫–∞–º–ø–∞–Ω–∏–π –≤ –∞–ø—Ä–µ–ª–µ‚Äì–º–∞–µ**.
- –¶–µ–Ω–æ–≤—É—é –ø–æ–ª–∏—Ç–∏–∫—É –º–æ–∂–Ω–æ –æ—Å—Ç–∞–≤–ª—è—Ç—å **—Å—Ç–∞–±–∏–ª—å–Ω–æ–π** ‚Äî —Å–µ–∑–æ–Ω–Ω–æ—Å—Ç—å —Å–ª–∞–±–æ –≤–ª–∏—è–µ—Ç –Ω–∞ —Ü–µ–Ω—É.

## –ó–∞–¥–∞—á–∞ 3. –ê–Ω–∞–ª–∏–∑ —Ä—ã–Ω–∫–∞ –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç–∏ –õ–µ–Ω–∏–Ω–≥—Ä–∞–¥—Å–∫–æ–π –æ–±–ª–∞—Å—Ç–∏

–û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –Ω–∞–∏–±–æ–ª–µ–µ –∞–∫—Ç–∏–≤–Ω—ã—Ö –∏ –ª–∏–∫–≤–∏–¥–Ω—ã—Ö –Ω–∞—Å–µ–ª—ë–Ω–Ω—ã—Ö –ø—É–Ω–∫—Ç–æ–≤ –≤ –õ–û: –≥–¥–µ –±–æ–ª—å—à–µ –≤—Å–µ–≥–æ –æ–±—ä—è–≤–ª–µ–Ω–∏–π –∏ –≤—ã—à–µ –¥–æ–ª—è –ø—Ä–æ–¥–∞–∂.

```python
# SQL-–∑–∞–ø—Ä–æ—Å, –≤—ã–ø–æ–ª–Ω–µ–Ω–Ω—ã–π –≤ DBeaver
sql_task3 = """
-- –û–ø—Ä–µ–¥–µ–ª–∏–º –∞–Ω–æ–º–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (–≤—ã–±—Ä–æ—Å—ã) –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–µ–π:
WITH limits AS (
    SELECT  
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY total_area) AS total_area_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY rooms) AS rooms_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY balcony) AS balcony_limit,
        PERCENTILE_DISC(0.99) WITHIN GROUP (ORDER BY ceiling_height) AS ceiling_height_limit_h,
        PERCENTILE_DISC(0.01) WITHIN GROUP (ORDER BY ceiling_height) AS ceiling_height_limit_l
    FROM real_estate.flats     
),
-- –ù–∞–π–¥—ë–º id –æ–±—ä—è–≤–ª–µ–Ω–∏–π, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ —Å–æ–¥–µ—Ä–∂–∞—Ç –≤—ã–±—Ä–æ—Å—ã:
filtered_id AS(
    SELECT id
    FROM real_estate.flats  
    WHERE 
        total_area < (SELECT total_area_limit FROM limits)
        AND (rooms < (SELECT rooms_limit FROM limits) OR rooms IS NULL)
        AND (balcony < (SELECT balcony_limit FROM limits) OR balcony IS NULL)
        AND ((ceiling_height < (SELECT ceiling_height_limit_h FROM limits)
            AND ceiling_height > (SELECT ceiling_height_limit_l FROM limits)) OR ceiling_height IS NULL)
), 
-- –ì–æ—Ç–æ–≤–∏–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è
categories AS (
	SELECT 
		fi.id AS id,
		days_exposition,
		last_price,
		last_price/total_area AS avg_pr_sm,
		total_area,
		TYPE,
		city,
		CASE
			WHEN days_exposition IS NULL 
			THEN '0'
			ELSE '1'
			END AS status
	FROM filtered_id AS fi
	INNER JOIN real_estate.flats AS f ON fi.id = f.id
	INNER JOIN real_estate.advertisement AS a ON a.id = f.id
	INNER JOIN real_estate.city AS c USING (city_id)
	INNER JOIN real_estate.TYPE AS t ON f.type_id = t.type_id
	WHERE city != '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥'
),
final_calc AS (
	SELECT 
		city,
		TYPE,
		count(status) AS ad_count, -- –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –æ–ø—É–±–ª–∏–∫–æ–≤–∞–Ω–Ω—ã—Ö –æ–±—ä—è–≤–ª–µ–Ω–∏–π
		sum(status::numeric) AS sales_count, -- –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–Ω—è—Ç—ã—Ö —Å –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏–π
		round(100*sum(status::numeric)/count(status)::NUMERIC,2) AS per_sales, -- –ø—Ä–æ—Ü–µ–Ω—Ç —Å–Ω—è—Ç—ã—Ö —Å –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏–π
		ROUND(AVG(avg_pr_sm::NUMERIC),0) AS avg_pr_sm, -- —Å—Ä–µ–¥–Ω—è—è —Ü–µ–Ω–∞ –∑–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –º–µ—Ç—Ä
		ROUND(AVG(total_area)::NUMERIC,2) AS avg_area, -- —Å—Ä–µ–¥–Ω—è—è –ø–ª–æ—â–∞—Ç—å –∫–≤–∞—Ä—Ç–∏—Ä—ã
		ROUND(avg(days_exposition)::numeric,0) AS avg_days_exp, -- —Å—Ä–µ–¥–Ω—è—è –ø—Ä–æ–¥–æ–ª–∂–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
		NTILE(4) OVER (ORDER BY avg(days_exposition) DESC) AS group_activ -- –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∞ –ø–æ —Å—Ä–µ–¥–Ω–µ–º—É –∑–Ω–∞—á–µ–Ω–∏—é –¥–Ω–µ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏
	FROM categories
	GROUP BY 
		city,
		TYPE
)
SELECT *
FROM final_calc
WHERE ad_count >= 50
ORDER BY 
	ad_count DESC;
"""

pd.read_csv("task3_result.csv", encoding="utf-8")
```

### üîπ –¶–µ–Ω–æ–≤–æ–π –¥–∏–∞–ø–∞–∑–æ–Ω ‚Äî —Ä–∞–∑–±—Ä–æ—Å –≤ 5.7 —Ä–∞–∑!

- **–ú–µ–¥–ª–µ–Ω–Ω–æ**: –ù–∏–∫–æ–ª—å—Å–∫–æ–µ (237 –¥–Ω–µ–π), –ö–æ–º–º—É–Ω–∞—Ä (236 –¥–Ω–µ–π)

- **–§–æ–∫—É—Å –Ω–∞ –ú—É—Ä–∏–Ω–æ, –®—É—à–∞—Ä—ã, –ö—É–¥—Ä–æ–≤–æ** ‚Äî –≤—ã—Å–æ–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å + —Å–∫–æ—Ä–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂.
- –í **–¥–æ—Ä–æ–≥–∏—Ö –ª–æ–∫–∞—Ü–∏—è—Ö** (–ü—É—à–∫–∏–Ω, –°–µ—Å—Ç—Ä–æ—Ä–µ—Ü–∫) ‚Äî –ø–æ–¥–∞–≤–∞—Ç—å **–ø—Ä–µ–º–∏—É–º-–ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ**.
- –í **–Ω–∏–∑–∫–æ–±—é–¥–∂–µ—Ç–Ω—ã—Ö** (–°–ª–∞–Ω—Ü—ã, –í–æ–ª—Ö–æ–≤) ‚Äî —Ñ–æ–∫—É—Å –Ω–∞ **—Å–æ—Ü–∏–∞–ª—å–Ω—É—é –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å**.
- –£—á–∏—Ç—ã–≤–∞—Ç—å **–ª–æ–∫–∞–ª—å–Ω—É—é —Å–∫–æ—Ä–æ—Å—Ç—å –ø—Ä–æ–¥–∞–∂**: –≤ –ù–∏–∫–æ–ª—å—Å–∫–æ–º ‚Äî –æ–∂–∏–¥–∞—Ç—å –¥–ª–∏–Ω–Ω—ã–π —Ü–∏–∫–ª.

## –û–±—â–∏–µ –≤—ã–≤–æ–¥—ã –∏ —Å—Ç—Ä–∞—Ç–µ–≥–∏—á–µ—Å–∫–∏–µ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏
